```python
import numpy as np
import matplotlib.pyplot as plt

# Definirea domeniului B
x = np.linspace(0.01, 0.99, 100)
y = np.linspace(0.01, 1, 100)
X, Y = np.meshgrid(x, y)
Z = X + 1j * Y

# Maparea utilizând funcția biliniară
f = (Z - 0.5 - 0.5j) / (Z + 0.5 - 0.5j)

# Afisarea graficului
plt.figure(figsize=(6, 6))
plt.scatter(Z.real, Z.imag, c='blue', alpha=0.5, label='Domeniul B')
plt.scatter(f.real, f.imag, c='red', alpha=0.5, label='Discul unitate')
plt.xlabel('Re(z)')
plt.ylabel('Im(z)')
plt.title('Mapare conformă a domeniului B pe discul unitate')
plt.legend()
plt.grid(True)
plt.show()

# Definirea domeniului H
x = np.linspace(-2, 2, 200)
y = np.linspace(-2, -0.01, 200)
X, Y = np.meshgrid(x, y)
Z = X + 1j * Y

# Maparea utilizând funcția biliniară
f = 1j * (Z - 1) / (Z + 1)

# Afisarea graficului
plt.figure(figsize=(6, 6))
plt.scatter(Z.real, Z.imag, c='blue', alpha=0.5, label='Domeniul H')
plt.scatter(f.real, f.imag, c='red', alpha=0.5, label='Semiplanul superior')
plt.xlabel('Re(z)')
plt.ylabel('Im(z)')
plt.title('Mapare conformă a domeniului H în semiplanul superior')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Definirea domeniului B
x = np.linspace(0.01, 0.99, 100)
y = np.linspace(0.01, 1, 100)
X, Y = np.meshgrid(x, y)
Z = X + 1j * Y

# Maparea utilizând funcția exponențială
f = np.exp(np.pi * (Z - 0.5j))

# Afisarea graficului
plt.figure(figsize=(6, 6))
plt.scatter(Z.real, Z.imag, c='blue', alpha=0.5, label='Domeniul B')
plt.scatter(f.real, f.imag, c='red', alpha=0.5, label='Semiplanul superior')
plt.xlabel('Re(z)')
plt.ylabel('Im(z)')
plt.title('Mapare conformă a domeniului B pe semiplanul superior')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Definirea domeniului H
x = np.linspace(-2, 2, 200)
y = np.linspace(-2, -0.01, 200)
X, Y = np.meshgrid(x, y)
Z = X + 1j * Y

# Maparea utilizând funcția tangenta hiperbolică
f = np.tanh(Z)

# Afisarea graficului
plt.figure(figsize=(6, 6))
plt.scatter(Z.real, Z.imag, c='blue', alpha=0.5, label='Domeniul H')
plt.scatter(f.real, f.imag, c='red', alpha=0.5, label='Discul unitate')
plt.xlabel('Re(z)')
plt.ylabel('Im(z)')
plt.title('Mapare conformă a domeniului H în discul unitate')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Definirea multimii Ω
x = np.linspace(0.01, 2, 100)
y = np.linspace(0.01, 2, 100)
X, Y = np.meshgrid(x, y)
Z = X + 1j * Y

# Aplicarea funcției e^z
W = np.exp(Z)

# Crearea graficului 3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, np.abs(W), cmap='jet')

# Setarea etichetelor și a titlului
ax.set_xlabel('Re(z)')
ax.set_ylabel('Im(z)')
ax.set_zlabel('|e^z|')
ax.set_title('Imaginea prin functia f(z) = e^z a multimii Ω')

# Afișarea graficului
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Definirea mulțimii Ω
x = np.linspace(0.01, 2, 400)
y = np.linspace(0.01, 2, 400)
X, Y = np.meshgrid(x, y)
Z = X + 1j * Y

# Aplicarea funcției e^z
W = np.exp(Z)

# Afișarea imaginii
plt.imshow(np.abs(W), extent=(0, 2, 0, 2), cmap='jet', origin='lower')
plt.colorbar(label='|e^z|')

# Setarea etichetelor și a titlului
plt.xlabel('Re(z)')
plt.ylabel('Im(z)')
plt.title('Imaginea prin funcția f(z) = e^z a mulțimii Ω')

# Afișarea plotului
plt.show()
```
