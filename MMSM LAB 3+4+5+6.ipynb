```python
#Ecuatii Diferentiale

import numpy as np
import matplotlib.pyplot as plt

def PunctCentral(f, t_initial, t_final, x_initial, N):
  t = np.linspace(t_initial, t_final, N)
  x = np.zeros(N)
  x[0] = x_initial
  h = t[1]-t[0]
  for i in range(N-1):
    k1 = h*f(t[i], x[i])
    k2 = h*f(t[i]+h/2, x[i]+k1/2)
    x[i+1] = x[i]+k2
  return x

t_initial = np.pi/2
t_final = 2*np.pi
x_initial = 1
f = lambda t, x: 2/t*x+t**2*np.cos(t)
N = 1000
fig = plt.figure(2)
ax = plt.axes()
ax.grid(True, color='black')
t_sol, x_sol = PunctCentral(f, t_initial, t_final, 1, 1000)
print(x_sol)
ax.plot(t_sol, x_sol, lw=3, color='r', label='x=x(t)')
ax.legend(loc = 'best')

```
